```python
class nn:
  def __init__(self,data,predictors,
               target,epochs=80,batch_size = 10):
    
    self.epochs, self.batch_size = epochs, batch_size
    self.df = data
    self.predictors = predictors
    self.target=target

    self.X, self.Y = None,None
    self.X_train, self.X_test = None, None
    self.Y_train, self.Y_test = None, None

    self.model = None

    self.predictions = None
    self.testingData = None

  def Data(self):
    X = self.df[self.predictors].values.tolist()

    self.Y = self.df[[self.target]].values.tolist()

    from sklearn.preprocessing import StandardScaler
    PredictorScaler=StandardScaler()
    
    # Storing the fit object for later reference
    self.PredictorScalerFit=PredictorScaler.fit(X)
    
    # Generating the standardized values of X and y
    self.X=self.PredictorScalerFit.transform(X)

    self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(self.X.tolist(), 
                                                                            self.Y, test_size=0.3, 
                                                                            random_state = 23, 
                                                                            shuffle = True)

  def Fit(self):

    self.model = Sequential()
    self.model.add(Dense(16, input_shape=(6,), activation='relu')) # Add an input shape! (features,)
    self.model.add(Dense(16, activation='relu'))
    self.model.add(Dense(1, activation='sigmoid'))
    # self.model.summary() 

    # compile the self.model
    self.model.compile(optimizer='Adam', 
                  loss='binary_crossentropy',
                  metrics=[metrics.categorical_accuracy])

    history = self.model.fit(self.X_train,
                        self.y_train,
                        # callbacks=[es],
                        epochs=self.epochs, # you can set this to a big number!
                        batch_size=self.batch_size,
                        # validation_split=0.2,
                        shuffle=True,
                        verbose=0)
    print('Fitted model w/ epochs:',self.epochs, 'batch size:',self.batch_size)
  # Defining a function for finding best hyperparameters
  def Predict(self):
    prob_str = self.target+'_probability'
    pred_str = self.target+'_prediction'
    self.predictions=self.model.predict(self.X_test)

    Test_Data=self.PredictorScalerFit.inverse_transform(self.X_test)

    # Generating a data frame for analyzing the test data
    self.testingData=pd.DataFrame(data=Test_Data, columns=self.predictors)
    self.testingData[self.target]=self.y_test
    self.testingData[prob_str]=self.predictions

    # Defining the probability threshold
    def probThreshold(inpProb):
      if inpProb > 0.5:
          return(1)
      else:
          return(0)

    # Generating predictions on the testing data by applying probability threshold
    self.testingData[pred_str]=self.testingData[prob_str].apply(probThreshold)
    print(self.testingData.head()) 
    print('\n######### Testing Accuracy Results #########')
    self.conf_matrix = skmetrics.confusion_matrix(self.y_test, self.testingData[pred_str])
    self.classification_report = skmetrics.classification_report(self.y_test, self.testingData[pred_str])
    self.f_score = skmetrics.f1_score(self.y_test, self.testingData[pred_str]) 
    print(self.conf_matrix)
    print(self.classification_report)
    return self.testingData   
  def Run(self):
    self.Data()
    self.Fit()
    return self.Predict()
  def Save(self):
    self.model.save('./models/'+self.target+'_model')
```

```python
def FunctionFindBestParams():
    
    # Defining the list of hyper parameters to try
    TrialNumber=0
    batch_size_list=[20]
    epoch_list=[100]
    
    import pandas as pd
    SearchResultsData=pd.DataFrame(columns=['TrialNumber', 'Parameters', 'f1_score'])
    
    for batch_size_trial in batch_size_list:
        for epochs_trial in epoch_list:
            TrialNumber+=1
            
            dummy_nn = nn(pwfDF,predictors,'PWF',epochs=epochs_trial,batch_size=batch_size_trial)
            dummy_nn.Run()

            # printing the results of the current iteration
            print(TrialNumber, 'Parameters:','batch_size:', batch_size_trial,'-', 'epochs:',epochs_trial, 'f1_score:', dummy_nn.f_score)
            
            SearchResultsData=SearchResultsData.append(pd.DataFrame(data=[[TrialNumber,
                            'batch_size'+str(batch_size_trial)+'-'+'epoch'+str(epochs_trial), dummy_nn.f_score]],
                                                                    columns=['TrialNumber', 'Parameters', 'f1_score'] ))
    return(SearchResultsData)
```
